# 引言

特殊函数类书籍是物理学家和工程师的核心数学工具。遗憾的是，这类书籍大多沿用 19 世纪的风格，内容主要围绕作为超几何类方程解的特殊函数展开。然而，在现代物理学的众多领域中，人们会用到更复杂的线性和非线性方程的解，因此如今也应将这类函数纳入特殊函数的范畴。例如，潘勒韦超越函数 —— 即潘勒韦非线性方程的解 —— 近年来已被公认为非线性物理学中的特殊函数。实际应用中，更复杂的原始方程及相应特殊函数的公式往往更为繁琐，若没有合适的指南，很难在文献中找到。另一方面，现代信息技术为文献查阅提供了新的方式，人们可以借助计算机工具，利用软件产品统一完成信息检索、交叉引用整理等工作，而非仅仅依赖书籍。

本书作者提出了一种将特殊函数理论视为数学、物理学与计算机技术交叉领域的研究方法。书中附带了读者可使用的软件产品，帮助读者借助计算机深入学习相关技术细节。本书中，严格的证明篇幅并不多，我们主要关注方程与特殊函数之间的各种关联，以便更好地理解该理论的精妙之处。我们还尽量通过直观的示例、表格和示意图来阐释理论，希望能助力读者更好地理解相关内容。

特殊函数源于微分方程、差分方程的解，也来自特定积分的计算过程以及群表示论等领域。在本书中，我们将把这些函数视为微分方程（更确切地说是二阶微分方程）的解来进行探讨。

若未在书中看到常规的方程历史介绍、标准符号及公认的公式推导，无需感到惊讶。这是因为本书基于新的结构要素，对这一知识领域采用了非传统的研究方法。下面列出其中的主要结构要素：
- 我们以奇异点的 s - 秩和微分方程的 s - 多重符号为基础，这些是对微分方程进行分类的有效工具。它们能将二阶常微分方程（ODE）划分为不同的类别和类型。这些概念在研究（非线性）潘勒韦方程时也十分有用。
- 方程的形式体现了多个等价方程的不同细节，实际上，它是另一个重要的结构要素。书中介绍了方程的标准形式：典范形式、正规形式和自共轭形式，其中自共轭形式在物理应用中经常出现。标准形式的公式比更复杂形式的方程公式更为简洁。
- 动态结构要素是方程的合流和约化过程。合流过程指两个奇点的融合，同时伴随着参数空间中的极限过渡。它能将原始方程解的性质推广到合流方程的解上。方程的约化过程是指在参数特殊化时改变方程类型的机制。如果方程的两个奇点在没有参数空间极限过渡的情况下融合，我们就称之为融合过程。

本书收录了作者近期的部分研究成果，其中包括新的渐近结果（S. Slavyanov，第 3 章）、雅菲展开的推广（V. Laine，第 1、3 章）、新的积分关系式（S. Slavyanov，第 3 章）、黎曼方案的推广（V. Laine、S. Slavyanov，第 1 章）、盖恩方程与潘勒韦方程之间的关联（S. Slavyanov，第 5 章）、矩阵元的递推关系式（S. Slavyanov，附录），详细参考文献可参见书末的文献列表。同时，书中所阐述的成果并非仅为作者个人研究所得，而是旨在满足那些在科学和工程研究中使用特殊函数作为工具的读者的需求。

本书的大部分内容是在巴登 - 符腾堡州完成的，这里也是伟大哲学家格奥尔格・威廉・弗里德里希・黑格尔的故乡。或许正因如此，许多结构要素都以三元组的形式呈现。我们先不给出具体定义（后续会有介绍），先列出其中最重要的一些：
- 超几何方程 —— 盖恩方程 —— 潘勒韦方程
- 富克斯方程 —— 合流方程 —— 约化合流方程
- 方程类别 —— 方程类型 —— 方程形式
- 微分方程 —— 差分方程 —— 积分方程
- 正则奇点 —— 非正则奇点 —— 非正则非分支奇点
- 局部解 —— 本征函数 —— 多项式解
- 局部参数 —— 标度参数 —— 辅助参数
- 合流过程 —— 约化过程 —— 融合过程

在本书撰写期间，圣彼得堡大学的两位年轻研究人员阿列克谢・阿科皮扬和阿列克谢・皮罗日尼科夫开发了配套软件 SFTools，其相关介绍见附录 B，获取该软件光盘的条件也可在附录 B 中找到。

对于这类包含大量显式公式且缺乏中间推导过程的书籍，出现排版错误几乎难以避免。但作者希望，若读者确实需要相关结果，能够自行验证推导过程；若仅需了解相关内容，也能对可能存在的错误予以包容。

关于本书的内容与结构，有几点说明：

- 第 1 章介绍具有多项式系数的二阶线性齐次微分方程及其解的一般理论。假定读者已掌握微分方程理论和复变函数论的基本概念。
- 接下来的两章分别讨论超几何类方程、特殊函数以及盖恩类方程、特殊函数。
- 第 4 章探讨特殊函数在物理学中的应用实例，作者尽量从不同领域选取示例。
- 第 5 章是潘勒韦方程理论的入门介绍。
- 附录中收集了一些未在正文中展开，但作者认为对理解全书内容有帮助的理论部分。

命题、证明和示例的表述均以符号□结尾。全书各处的 ε 均表示任意小的正数。
